```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Build a book recommender system 
# data source: http://www2.informatik.uni-freiburg.de/~cziegler/BX/
```

```python
books = pd.read_csv('data/BX-Books.csv', sep=';', error_bad_lines=False, encoding="latin-1", warn_bad_lines=False) 
# "error_bad_lines" is used to handle bad lines (too many/few fields) 
print list(books)
print books.size
books.sample(5)
```

```python
users = pd.read_csv('data/BX-Users.csv', sep=';', error_bad_lines=False, encoding="latin-1")
users.columns = ['userID', 'Location', 'Age']
users.sample(5)
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")
ratings.columns = ['userID', 'ISBN', 'bookRating']
ratings.sample(5)
```
